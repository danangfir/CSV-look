```python
import pandas as pd
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output
import io

# Membuat widget untuk unggah file
upload_widget = widgets.FileUpload(
    accept='data_kesejahteraan.csv',
    multiple=False
)

# Menampilkan widget unggah file
display(upload_widget)
```

```python
# Menunggu file diunggah dan memuatnya ke dalam dataframe
uploaded_file = list(upload_widget.value.values())[0]
file_content = uploaded_file['content']

# Membaca file CSV yang diunggah
df = pd.read_csv(io.BytesIO(file_content))
print("Data Awal:")
display(df.head())
```

```python
# Mengecek apakah ada nilai yang hilang
missing_values = df.isnull().values.any()
print("\nApakah ada nilai yang hilang di dataset?:", missing_values)
```

```python
# Menampilkan jumlah nilai yang hilang di setiap kolom
print("\nJumlah nilai yang hilang di setiap kolom:")
display(df.isnull().sum())
```

```python
# Mengisi missing values di kolom numerik dengan rata-rata
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())
print("\nMengisi missing values dengan rata-rata (untuk kolom numerik)...")
```

```python
# Mengisi missing values di kolom teks dengan modus
text_cols = df.select_dtypes(include='object').columns
for column in text_cols:
    if df[column].isnull().any():
        df[column].fillna(df[column].mode()[0], inplace=True)
print("Mengisi missing values dengan modus (untuk kolom teks)...")
```

```python
# Mengecek kembali apakah masih ada missing values
print("\nJumlah nilai yang hilang setelah penanganan:")
display(df.isnull().sum())
```

```python
# Menyimpan dataframe yang sudah diproses ke CSV baru
df.to_csv('data_kesejahteraan_bersih.csv', index=False)
print("\nData telah disimpan sebagai 'data_kesejahteraan_bersih.csv'.")
```
